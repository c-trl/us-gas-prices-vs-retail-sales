#When prices at the pump go down, are consumers more inclined to spend?  

This exercise looks at average national gas prices and real retail and food service sales 

```python
import pandas as pd
from pandas import *
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('usgas.csv')
retail = pd.read_csv('https://raw.githubusercontent.com/c-trl/us-gas-prices-vs-retail-sales/master/retailsales.csv')

#date range to use: 1992-01 - 2014-11
```

```python
#trimming gas to match retail date range
gas = gas.drop(gas.index[[arange(0,16)]]).drop(gas.tail(2).index).reset_index()

#resetting index after dropping rows
gas = gas.drop('index',1)

#merging gas data and retail sales data
df = merge(gas,retail,on='Month',how='inner')
```

```python
df.plot(subplots=True, figsize=(15, 10), xlim=(0,274), color='#111111', linewidth=5);
```

After looking at the plotted data side-by-side, <b>it's hard to say there is any significant instance of causality</b> here.

From the graphs alone, we CAN see the difference between the effect of the Great Recession on retail sales and gas prices.  There are also a number of instances where, <b>albeit slightly, we can see a delayed increase in retail sales following small dips in gas prices</b>.

Perhaps a better way to see a comparison in trends would be to look at <b>%-change</b> numbers.
